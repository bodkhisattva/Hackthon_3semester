```python
from sklearn.svm import LinearSVC
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
class MLMailClassifier:
    def __init__(self):
        self.vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2), sublinear_tf=True)
        self.model = LinearSVC()
    
    def train(self, texts, labels):
        X = self.vectorizer.fit_transform(texts)
        self.model.fit(X, labels)
    
    def predict(self, text, top_n=3):
        X = self.vectorizer.transform([text])
        scores = self.model.decision_function(X)
        if len(scores.shape) == 1:
            scores = [scores]
        scores = scores[0]
        categories = self.model.classes_
        top_indices = scores.argsort()[::-1][:top_n]
        result = [(categories[i], float(scores[i])) for i in top_indices]
        if scores[top_indices[0]] < 0.1:
            return [("Не определена", scores[top_indices[0]])]
        return result
```
